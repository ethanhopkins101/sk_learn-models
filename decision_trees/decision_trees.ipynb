```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier as dtc , plot_tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
x
```

### creating Decision tree

```python
clf=dtc()
```

```python
clf=clf.fit(x,y)
```

```python
a=np.array([[6.7, 3. , 5.2, 2.3],
       [6.3, 2.5, 5. , 1.9]])
```

```python
y_predict=clf.predict(a)
```

```python
y_predict
```

### Visualizing Tree

```python
viz=plot_tree(clf)
```

```python
plt.figure(figsize=(12,8))
plot_tree(clf,filled=True)
plt.show()
```
