```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report as cr, ConfusionMatrixDisplay as cmd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score
```

```python
data= load_iris()
inputs= data.data
target= data.target
```

```python
#splitting
x_train, x_test, y_train, y_test= tts(inputs, target, test_size= .3, random_state=22, stratify= target)
```

```python
#model tree:
dtc= DecisionTreeClassifier(ccp_alpha= 0.1)
dtc.fit(x_train, y_train)
y_pred= dtc.predict(x_test)
```

```python
# model random forrest:
rfc= RandomForestClassifier()
rfc.fit(x_train, y_train)
y_pred_rfc= rfc.predict(x_test)
```

```python
plt.figure(figsize= (12,6))
plot_tree(dtc, filled= True)
plt.show()
```

```python
print(cr(y_test, y_pred))
```

```python
fig, ax= plt.subplots(1, 2, figsize= (12,6))
cmd.from_predictions(y_test, y_pred, cmap= 'BuPu_r', ax= ax[0])
cmd.from_predictions(y_test, y_pred_rfc, cmap= 'BuPu_r', ax= ax[1])
plt.show()
```
